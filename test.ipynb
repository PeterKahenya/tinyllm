```python
import os
import base64
import json
```

```python
with open('cl100k_base.tiktoken', 'r') as f:
    base = f.read()
base
```

```python
def load_tiktoken_bpe(contents) -> dict[bytes, int]:
    return {
        base64.b64decode(token): int(rank)
        for token, rank in (line.split() for line in contents.splitlines() if line)
    }
```

```python
encoder = load_tiktoken_bpe(base)
encoder
```

```python
encoder = {k.decode('latin'): v for k, v in encoder.items()}
```

```python
# save the cl100k_base.tiktoken as a json file
with open('cl100k_base.tiktoken.json', 'w') as f:
    json.dump(encoder, f)
```
